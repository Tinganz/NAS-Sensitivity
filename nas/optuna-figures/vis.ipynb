```python
from pathlib import Path
from IPython.display import display
import ipywidgets as widgets
import plotly.graph_objects as go
import re, json

FIG_DIR = Path.cwd()
html_files = sorted(FIG_DIR.glob('*.html'))

def extract_fig(path):
    m = re.search(
        r'Plotly\.newPlot\(\s*["\'].*?["\']\s*,\s*(\[.*?\])\s*,\s*(\{.*?\})\s*,\s*(\{.*?\})\s*\)',
        Path(path).read_text(), re.DOTALL
    )
    return go.Figure(data=json.loads(m.group(1)), layout=json.loads(m.group(2)))

dropdown = widgets.Dropdown(options=[(p.name, str(p)) for p in html_files], description='Figure:')
out = widgets.Output()

def render_fig(change):
    out.clear_output(wait=True)
    with out:
        extract_fig(change['new']).show()

dropdown.observe(render_fig, names='value')
display(dropdown, out)
render_fig({'new': dropdown.value})
```
